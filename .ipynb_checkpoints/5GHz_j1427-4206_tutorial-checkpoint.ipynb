## 5.2 <a name="section5.2">Fit a baseline to remove systemic contributions</a>
[back to top](#top)

We are only interested in the temperature contributed by the source, so we want to subract off the system temperature contribution.  We see that the baseline level changes through during the scan.

What do we fit then?  This is a fairly strong source, we can clealy see the antenna beam pattern.  Looking carefully we can see a deflection point at the base of the beam.  This is the first null of the beam. At these points we will not be receiving any power from the source.


Locate the section you want to use for your fit. We are going to write a few methods to help us process the location selection for our fit

```python

def setBaseSize(len_scan):
    """
        Select the size of your baseline based on a fraction of the 
        total length of the scan.
    """
    
    len_size_factor = 0.04 #use  6% of data points for fitting baseline
    minPts = int(len_scan*len_size_factor)
    basePts = minPts*2

    return basePts

def setBasePts(x,y,len_scan): 
    '''
        Set baseline points. Select points along the driftscan where you will fit
        your baseline.
    '''

    basePtsLeft  = setBaseSize(len_scan)
    basePtsRight = len_scan - basePtsLeft 

    xbleft = x[:basePtsLeft]
    ybleft = y[:basePtsLeft]
    xbright= x[basePtsRight:]
    ybright= y[basePtsRight:]

    xbasepts = list(xbleft) + list(xbright)
    ybasepts = list(ybleft) + list(ybright)

    return xbasepts, ybasepts 
```

Fit the baseline and plot your data

```python

def fitBasePts(xb,yb,ra, drift):
    """Fit the data to remove systematic contributions.
    """
        
    basefitcoeffs = np.polyfit(xb, yb, 1)
    basefitline = np.polyval(basefitcoeffs,xb)   
    datafit = np.polyval(basefitcoeffs,ra)
    datafit = drift - datafit
    
    return datafit, basefitline
    
def fitBaselineAndPlot(ra,drift,title):
    """
        
    """

    # Get the length of the scans
    lenscan = len(ra)

    # Find the baseline points
    xb,yb = setBasePts(ra,drift,lenscan)
    
    # Fit the baseline points
    fit,bf = fitBasePts(xb,yb, ra, drift)

    # plot the data
    plt.figure(figsize=[15,5])
    plt.suptitle(title)

    ax = plt.subplot(121)
    plt.axhline(y=0, color='k', alpha= 0.5)
    l1,= plt.plot(ra,drift, label = 'original data')
    l2,= plt.plot(xb,yb,'r.', label = 'noise')
    plt.plot(xb,bf, 'g', label = "baseline fit")
    plt.xlabel('RA offset (deg)')
    plt.ylabel('Ta (K)')
    plt.legend()
    plt.grid()
    lim = plt.axis('tight')

    ax = plt.subplot(122)
    l1,= plt.plot(ra,fit, label = 'final data')
    plt.xlabel('RA offset (deg)')
    plt.ylabel('Ta (K)')
    plt.axhline(y=0, color='k', alpha= 0.5)
    plt.legend()
    plt.grid()
    lim = plt.axis('tight')
    
    return fit

# Fit and plot 

slcp  = fitBaselineAndPlot(ra, south_scan_LCP, 'South offset scan [LCP]')
srcp  = fitBaselineAndPlot(ra, south_scan_RCP, 'South offset scan [RCP]')
nlcp  = fitBaselineAndPlot(ra, north_scan_LCP, 'North offset scan [LCP]')
nrcp  = fitBaselineAndPlot(ra, north_scan_RCP, 'North offset scan [RCP]')
onlcp = fitBaselineAndPlot(ra, on_scan_LCP, 'On scan [LCP]')
onrcp = fitBaselineAndPlot(ra, on_scan_RCP, 'On scan [RCP]')
```

Hydra A is partially resolved by the 26-m telescope, so what we see is not a perfect beam pattern. We don't get a perfect fit, but it can help us disentangle the various components and zoom in on an area in which we can do a  more specialised fit.

We have good estimates for the baseline as shown in the figures above.

## 5.3 <a name="section5.3">Fit beam to get antenna temperature</a>
[back to top](#top)


We are now going to fit the top of the beam to fit the 
antenna temperature. We will use a 2nd order polynomial fit
at the top of the peak. 

```python
# Get location of peak
def getAPeakPts(drift):
    """Get locations of scan peak where we are going to fit 
    """
    peak_max = max(drift)
    peak_pts = np.where(drift > 0.6*peak_max)[0]
    return peak_pts

def getBPeakPts(drift):
    """Get locations of scan peak where we are going to fit 
    """
    peak_min = max(drift)
    peak_pts = np.where(drift < -0.6*peak_min)[0]
    return peak_pts

# Get the residual and rms to estimate the errors in the fit
def residual(model, data): 
    
    from sklearn.metrics import mean_squared_error
    """
        Calculate the residual between the model and the data.
    """

    res = np.array(model - data)
    rms = np.sqrt(mean_squared_error(data,model))

    return res, rms

# Fit the peak
def fitPeak(peak_pts, ra, drift):
    """Fit the peak.
    """

    peakfitcoeffs = np.polyfit(ra[peak_pts], drift[peak_pts], 2)
    peakfitline = np.polyval(peakfitcoeffs,ra[peak_pts])
    
    res, rms = residual(peakfitline, drift[peak_pts])
    return peakfitline,rms

# Get peak points, Abeam
Apeak_pts_slcp= getAPeakPts(slcp)
Apeak_pts_srcp = getAPeakPts(srcp)
Apeak_pts_nlcp = getAPeakPts(nlcp)
Apeak_pts_nrcp = getAPeakPts(nrcp)
Apeak_pts_onlcp = getAPeakPts(onlcp)
Apeak_pts_onrcp = getAPeakPts(onrcp)

# Get peak points, Bbeam
Bpeak_pts_slcp= getBPeakPts(slcp)
Bpeak_pts_srcp = getBPeakPts(srcp)
Bpeak_pts_nlcp = getBPeakPts(nlcp)
Bpeak_pts_nrcp = getBPeakPts(nrcp)
Bpeak_pts_onlcp = getBPeakPts(onlcp)
Bpeak_pts_onrcp = getBPeakPts(onrcp)

# Fit the peaks
Afitpsl, Aslrms  = fitPeak(Apeak_pts_slcp, ra, slcp)
Afitpsr, Asrrms = fitPeak(Apeak_pts_srcp, ra, srcp)
Afitpnl, Anlrms = fitPeak(Apeak_pts_nlcp, ra, nlcp)
Afitpnr, Anrrms = fitPeak(Apeak_pts_nrcp, ra, nrcp)
Afitponl, Aonlrms = fitPeak(Apeak_pts_onlcp, ra, onlcp)
Afitponr, Aonrrms = fitPeak(Apeak_pts_onrcp, ra, onrcp)

Bfitpsl, Bslrms  = fitPeak(Bpeak_pts_slcp, ra, slcp)
Bfitpsr, Bsrrms = fitPeak(Bpeak_pts_srcp, ra, srcp)
Bfitpnl, Bnlrms = fitPeak(Bpeak_pts_nlcp, ra, nlcp)
Bfitpnr, Bnrrms = fitPeak(Bpeak_pts_nrcp, ra, nrcp)
Bfitponl, Bonlrms = fitPeak(Bpeak_pts_onlcp, ra, onlcp)
Bfitponr, Bonrrms = fitPeak(Bpeak_pts_onrcp, ra, onrcp)

# Print out the ANTENNA TEMPERATURES
print 'Ta_ASLCP = %.3f +- %.3f [K], Ta_BSLCP = %.3f +- %.3f [K]' %(max(Afitpsl), Aslrms, min(Bfitpsl), Bslrms)
print 'Ta_ASRCP = %.3f +- %.3f [K], Ta_BSRCP = %.3f +- %.3f [K]' %(max(Afitpsr), Asrrms, min(Bfitpsr), Bsrrms)
print 'Ta_ANLCP = %.3f +- %.3f [K], Ta_BNLCP = %.3f +- %.3f [K]' %(max(Afitpnl), Anlrms, min(Bfitpnl), Bnlrms)
print 'Ta_ANRCP = %.3f +- %.3f K], Ta_BNRCP = %.3f +- %.3f K]' %(max(Afitpnr), Anrrms, min(Bfitpnr), Bnrrms)
print 'Ta_AONLCP = %.3f +- %.3f [K], Ta_BONLCP = %.3f +- %.3f [K]' %(max(Afitponl), Aonlrms, min(Bfitponl), Bonlrms)
print 'Ta_AONRCP = %.3f +- %.3f [K], Ta_BONRCP = %.3f +- %.3f [K]' %(max(Afitponr), Aonrrms, min(Bfitponr), Bonrrms)
```

## 5.4 <a name="section5.4">Amplitude correction</a>
[back to top](#top)

As you can see our pointing is off and we need to correct for that

```python
def correctAmp(hpn,errhpn,hps,errhps,on,erron):

    '''Calculate the pointing correction
    '''

    if on != 0.0:
        
        #if no HPN
        if((hpn == 0) and (hps != 0)):
            '''
                Tcorr = Ton x exp[((ln(Thps) - ln(Ton) + ln(2))**2)/term2]
            '''

            term1 = (np.log(abs(hps)) - np.log(abs(on)) - np.log(2))**2
            term2 = 4*np.log(2)
            term3 = term1/term2
            pc = np.exp(term3)
            Tcorr = on * pc

            #calculate the error
            if hps < 0.0:
                pf1 = pc * 2.0 * term3 * (-1.0/hps)
            else:
                pf1 = pc * 2.0 * term3 * (1.0/hps)
                                
            pf2 = pc * 2.0 * term3 * (1.0/on)
            Tcorerr = np.sqrt((erron**2)*(pf2**2) + (errhps**2)*(pf1**2))
            
        #if no HPS
        elif((hpn != 0) and (hps== 0)):
            '''
                Tcorr = Ton x pc

                term1 = ((ln(Ton) - ln(Thpn) - ln(2))**2)
                term2 = (4*ln(2))
                term3 = term1/term2
                pc = exp[term3]
            '''

            term1 = (np.log(abs(on)) - np.log(abs(hpn)) - np.log(2))**2
            term2 = 4*np.log(2)
            term3 = term1/term2
            pc = np.exp(term3)
            Tcorr = on * pc

            #calculate the error
            if hpn < 0.0:
                pf1 = pc * 2.0 * term3 * (-1.0/hpn)
            else:
                pf1 = pc * 2.0 * term3 * (1.0/hpn)

            pf2 = pc * 2.0 * term3 * (1.0/on)
            Tcorerr = np.sqrt((erron**2)*(pf2**2) + (errhpn**2)*(pf1**2))


        #if all present
        elif((hpn != 0) and (hps!= 0)):
            '''
                Tcorr = Ton x exp[((ln(Thps) - ln(Thpn))**2)/(16*ln(2))]
            '''
            '''
                Tcorr = Ton x pc

                term1 = ((ln(Thps) - ln(Thpn))**2)
                term3 = term1/term2
                pc = exp[term3]
            '''

            term1 = (np.log(abs(hps)) - np.log(abs(hpn)))**2
            term2 = 16*np.log(2)
            term3 = term1/term2
            pc = np.exp(term3)
            
            Tcorr = abs(on * pc)

            #calculate the error in the 
            
            if hps < 0.0:
                pf1 = pc * 2.0 * term3 * (1.0/hps)
            else:
                pf1 = pc * 2.0 * term3 * (1.0/abs(hps))
                                
            if hpn < 0.0:
                pf2 = pc * 2.0 * term3 * (-1.0/hpn)
            else:
                pf2 = pc * 2.0 * term3 * (1.0/abs(hpn))
            
            Tcorerr = np.sqrt((errhpn**2)*(pf2**2) + (errhps**2)*(pf1**2))

            '''print pc
            print Tcorr
            print Tcorerr, '\n'
            '''
            
        #if no hpn/hps
        if((hpn == 0) and (hps== 0)):
            '''
                Tcorr = 0
            '''

            pc=0
            Tcorr=0
            Tcorerr=0
    else:
        pc=0
        Tcorr=0
        Tcorerr=0
            
    return pc, Tcorr, Tcorerr
```

Calculate the corrected amplitude



```python
# Calculate Pointing correction
Apcl, ATcorrl, ATcorerrl = correctAmp(max(Afitpnl), Anlrms, max(Afitpsl), Aslrms, max(Afitponl), Aonlrms)
Bpcl, BTcorrl, BTcorerrl = correctAmp(min(Bfitpnl), Bnlrms, min(Bfitpsl), Bslrms, min(Bfitponl), Bonlrms)

Apcr, ATcorrr, ATcorerrr = correctAmp(max(Afitpnr), Anrrms, max(Afitpsr), Asrrms, max(Afitponr), Aonrrms)
Bpcr, BTcorrr, BTcorerrr = correctAmp(min(Bfitpnr), Bnrrms, min(Bfitpsr), Bsrrms, min(Bfitponr), Bonrrms)


print 'Amplitude correction factor due to pointing error in A BEAM LCP = %.3f ' %(Apcl)
print 'Amplitude correction factor due to pointing error in B BEAM LCP = %.3f ' %(Bpcl)
print 'Amplitude correction factor due to pointing error in A BEAM RCP = %.3f ' %(Apcr)
print 'Amplitude correction factor due to pointing error in B BEAM RCP = %.3f ' %(Bpcr)

print 'Ta_ALCP corr = %.3f +- %.3f [K]' %(ATcorrl, ATcorerrl)
print 'Ta_BLCP corr = %.3f +- %.3f [K]' %(BTcorrl, BTcorerrl)
print 'Ta_ARCP corr = %.3f +- %.3f [K]' %(ATcorrr, ATcorerrr)
print 'Ta_BRCP corr = %.3f +- %.3f [K]' %(BTcorrr, BTcorerrr)
```

# 6. <a name="section6">Calculate the Flux density</a>

```python
def fluxDensity(pss,errpss,ta,errta):

    '''Calculate the source flux density'''
    
    flux = pss*ta
    errflux = np.sqrt((errpss**2)/(pss**2) + (errta**2)/(ta**2)) * flux

    return flux, errflux

def totfluxDensity(Afluxlcp,Aerrfluxlcp,Bfluxlcp,Berrfluxlcp,Afluxrcp,Aerrfluxrcp,Bfluxrcp,Berrfluxrcp):

    '''Calculate the source flux density'''
    
    flux = ((Afluxlcp + Bfluxlcp + Afluxrcp + Bfluxrcp)/4.0)*2
    errflux = np.sqrt((Aerrfluxlcp**2)/(Afluxlcp**2) + (Berrfluxlcp**2)/(Bfluxlcp**2) + (Aerrfluxrcp**2)/(Afluxrcp**2) + (Berrfluxrcp**2)/(Bfluxrcp**2) )*flux

    return flux, errflux
```

```python
# Calculate flux density

#PSS @ 5 GHz

Apsslcp = 6.03
Aerrpsslcp = 0.0
Bpsslcp = 6.05
Berrpsslcp = 0.0
Apssrcp = 6.47
Aerrpssrcp =  0.0
Bpssrcp = 6.34
Berrpssrcp = 0.0


#Flux
Afluxlcp, Aerrfluxlcp = fluxDensity(Apsslcp,Aerrpsslcp,ATcorrl,ATcorerrl)
Bfluxlcp, Berrfluxlcp = fluxDensity(Bpsslcp,Berrpsslcp,BTcorrl,BTcorerrl)
Afluxrcp, Aerrfluxrcp = fluxDensity(Apssrcp,Aerrpssrcp,ATcorrr,ATcorerrr)
Bfluxrcp, Berrfluxrcp = fluxDensity(Bpssrcp,Berrpssrcp,BTcorrr,BTcorerrr)


totflux, errtotflux = totfluxDensity(Afluxlcp, Aerrfluxlcp,Bfluxlcp, Berrfluxlcp,Afluxrcp, Aerrfluxrcp,Bfluxrcp, Berrfluxrcp)

#print 'Flux_LCP corr = %.3f +- %.3f [K]' %(fluxlcp, errfluxlcp)
#print 'Flux_RCP corr = %.3f +- %.3f [K]' %(fluxrcp, errfluxrcp)
print 'Total_flux = %.3f +- %.3f [K]' %(totflux, errtotflux)

```
