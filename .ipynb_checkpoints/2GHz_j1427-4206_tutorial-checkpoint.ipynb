## Fit a baseline to remove systemic contributions

We are only interested in the temperature contributed by the source, so we want to subract off the system temperature contribution.  We see that the baseline level changes through during the scan, and the scan also crosses the first sidelobes.  

What do we fit then?  This is a fairly strong source, we can clealy see the antenna beam pattern.  Looking carefully we can see a deflection point at the base of the beam.  This is the first null of the beam. At these points we will not be receiving any power from the source.

The second table in the fits file gives us useful information on the 2.5cm feed system, including information about the beam.

```python
#hdulist[1].header
```

We will use the first null beam width points to set baseline points for the noise level and the half power beam width for the peak points. 

```python
hpbw = hdulist[1].header['HPBW']          #half power beamwidth
hfnbw = hdulist[1].header['FNBW']/2.0     #half beamwidth to first nulls
```

```python
def setNoiseLevel(fnbw, offset):
    """Use the first null beam width to set the baseline
       noise level.
    """
    noise_left  = np.where((offset <= -1*fnbw))[0]
    noise_right = np.where((offset >= fnbw))[0]
    noise_level_points = np.concatenate((noise_left,noise_right))
    return noise_level_points
        
def fitBaseline(noise, offset, drift):
    """Fit the data to remove systematic contributions.
    """

    basefitcoeffs = np.polyfit(offset[noise], drift[noise], 1)
    basefitline = np.polyval(basefitcoeffs,offset[noise])
    
    #print basefitcoeffs
    datafit = np.polyval(basefitcoeffs,offset)
    datafit = drift-datafit
    return datafit,basefitline
    
```

```python
def fitBaseAndPlot(offset,drift,title):
    # Find the locations of the sytem noise level
    noise = setNoiseLevel(hfnbw, offset)
    fit,bf = fitBaseline(noise, offset, drift)

    # plot the data
    plt.figure(figsize=[15,5])
    plt.suptitle(title)

    ax = plt.subplot(121)
    plt.axvline(x=0, color='k', alpha=0.5)
    plt.axhline(y=0, color='k', alpha= 0.5)
    l1,= plt.plot(offset,drift, label = 'original data')
    l2,= plt.plot(offset[noise],drift[noise],'r.', label = 'noise')
    plt.plot(offset[noise],bf, 'g', label = "baseline fit")
    plt.xlabel('RA offset (deg)')
    plt.ylabel('Ta (K)')
    plt.legend()
    plt.grid()
    lim = plt.axis('tight')

    ax = plt.subplot(122)
    l1,= plt.plot(offset,fit, label = 'final data')
    #l2,= plt.plot(offset[noise],fit[noise],'r.', label = 'noise')
    plt.xlabel('RA offset (deg)')
    plt.ylabel('Ta (K)')
    plt.axvline(x=0, color='k', alpha=0.5)
    plt.axhline(y=0, color='k', alpha= 0.5)
    plt.legend()
    plt.grid()
    lim = plt.axis('tight')
    
    return fit
```

```python
# Fit and plot 

slcp  = fitBaseAndPlot(offset, south_scan_LCP, 'South offset scan [LCP]')
srcp  = fitBaseAndPlot(offset, south_scan_RCP, 'South offset scan [RCP]')
nlcp  = fitBaseAndPlot(offset, north_scan_LCP, 'North offset scan [LCP]')
nrcp  = fitBaseAndPlot(offset, north_scan_RCP, 'North offset scan [RCP]')
onlcp = fitBaseAndPlot(offset, on_scan_LCP, 'On scan [LCP]')
onrcp = fitBaseAndPlot(offset, on_scan_RCP, 'On scan [RCP]')
```

Hydra A is partially resolved by the 26-m telescope, so what we see is not a perfect beam pattern. We don't get a perfect fit, but it can help us disentangle the various components and zoom in on an area in which we can do a  more specialised fit.

We have good estimates for the baseline as well as the position of the first nulls as shown in the figures above.

## Fit beam to get antenna temperature

We are now going to fit the top of the beam to fit the 
antenna temperature. We will use a 2nd order polynomial fit
at the top 20% of the peak to estimate the height of the 
peak. 

```python
#Get indices where the offset values are within the main beam
#main_beam = np.where(np.logical_and(offset >= -hpbw, offset <= hpbw))[0]

#Get indices for peak 20%
def getPeakPts(drift):
    """Get locations of scan peak where we are going to fit 
    """
    pmax = max(drift)
    peak_pts = np.where(drift > 0.5*pmax)[0]
    return peak_pts

# Get the residual and rms
def residual(model, data): 
    from sklearn.metrics import mean_squared_error
    """
        Calculate the residual between the model and the data.
    """

    res = np.array(model - data)
    rms = np.sqrt(mean_squared_error(data,model))

    return res, rms

# Fit the peak
def fitPeak(peak_pts, offset, drift):
    """Fit the data to remove systematic contributions.
    """

    peakfitcoeffs = np.polyfit(offset[peak_pts], drift[peak_pts], 2)
    peakfitline = np.polyval(peakfitcoeffs,offset[peak_pts])
    
    res, rms = residual(peakfitline, drift[peak_pts])
    return peakfitline,rms

# Get peak points
peak_pts_slcp= getPeakPts(slcp)
peak_pts_srcp = getPeakPts(srcp)
peak_pts_nlcp = getPeakPts(nlcp)
peak_pts_nrcp = getPeakPts(nrcp)
peak_pts_onlcp = getPeakPts(onlcp)
peak_pts_onrcp = getPeakPts(onrcp)

# Fit the peaks
fitpsl, slrms  = fitPeak(peak_pts_slcp, offset, slcp)
fitpsr, srrms = fitPeak(peak_pts_srcp, offset, srcp)
fitpnl, nlrms = fitPeak(peak_pts_nlcp, offset, nlcp)
fitpnr, nrrms = fitPeak(peak_pts_nrcp, offset, nrcp)
fitponl, onlrms = fitPeak(peak_pts_onlcp, offset, onlcp)
fitponr, sonrrms = fitPeak(peak_pts_onrcp, offset, onrcp)

print 'Ta_SLCP = %.3f +- %.3f [K]' %(max(fitpsl), slrms)
print 'Ta_SRCP = %.3f +- %.3f [K]' %(max(fitpsr), srrms)
print 'Ta_NLCP = %.3f +- %.3f [K]' %(max(fitpnl), nlrms)
print 'Ta_NRCP = %.3f [+- %.3f K]' %(max(fitpnr), nrrms)
print 'Ta_ONLCP = %.3f +- %.3f [K]' %(max(fitponl), onlrms)
print 'Ta_ONRCP = %.3f +- %.3f [K]' %(max(fitponr), sonrrms)


#np.where(slcp >= max(fitpsl))[0]
```

```python
def plotFit(offset, drift1, title, peak_pts1,fit1,drift2, peak_pts2,fit2, lab1,lab2):
    # plot the data
    plt.figure(figsize=[15,5])
    plt.suptitle(title)

    ax = plt.subplot(121)
    l1,= plt.plot(offset,drift1, label = 'original data')
    #l2,= plt.plot(offset[peak_pts1],drift1[peak_pts1])
    l3,= plt.plot(offset[peak_pts1],fit1,'r', label=lab1)
    plt.xlabel('RA offset (deg)')
    plt.ylabel('Ta (K)')
    plt.legend()
    plt.axvline(x=0, color='k', alpha=0.5)
    plt.axhline(y=0, color='k', alpha= 0.5)
    plt.grid()
    lim = plt.axis('tight')

    ax = plt.subplot(122)
    l1,= plt.plot(offset,drift2, label = 'original data')
    #l2,= plt.plot(offset[peak_pts2],drift2[peak_pts2])
    l3,= plt.plot(offset[peak_pts2],fit2, 'r',label = lab2)
    plt.xlabel('RA offset (deg)')
    plt.ylabel('Ta (K)')
    plt.axvline(x=0, color='k', alpha=0.5)
    plt.axhline(y=0, color='k', alpha= 0.5)
    plt.legend()
    plt.grid()
    lim = plt.axis('tight')

plotFit(offset,slcp,'HPS PEAK FIT',peak_pts_slcp,fitpsl, srcp, peak_pts_srcp,fitpsr,'Ta_SLCP = %.3f +- %.3f [K]' %(max(fitpsl), slrms), 'Ta_SRCP = %.3f +- %.3f [K]' %(max(fitpsr), srrms))
plotFit(offset,nlcp,'HPN PEAK FIT',peak_pts_nlcp,fitpnl, nrcp, peak_pts_nrcp,fitpnr, 'Ta_NLCP = %.3f +- %.3f [K]' %(max(fitpnl), nlrms), 'Ta_NRCP = %.3f [+- %.3f K]' %(max(fitpnr), nrrms))
plotFit(offset,onlcp,'ON PEAK FIT',peak_pts_onlcp,fitponl, onrcp, peak_pts_onrcp,fitponr,'Ta_ONLCP = %.3f +- %.3f [K]' %(max(fitponl), onlrms), 'Ta_ONRCP = %.3f +- %.3f [K]' %(max(fitponr), sonrrms))

```

The old data reduction program at HartRAO would fit a 2nd order polynomial through the top of the peak, because the beam isn't really a Gaussian - its actually a modified sin(x)/x function.
Fitting s 2nd order polynomial gives us a reasonable fit so we can continue.

Well,that seems reasonable enough - your guess for the linear component can be outrageous and the curve fitter will still converge, but your guess for the centre of the gaussian should be at least somewhere inside the main beam. We are already making the assumption that the first nulls are covered by the scan (ie the pointing model isn't terribly off) and so we are probably ok with assuming that the centre of the source is near the centre of the scan but system temperature can vary considerably so its good that the fit is not sensitive to the guess at the baseline level.

# 6. Calculate the Pointing correction

```python
def calcPC(hpn,errhpn,hps,errhps,on,erron):

    '''Calculate the pointing correction
    '''

    if on != 0.0:
        
        #if no HPN
        if((hpn == 0) and (hps != 0)):
            '''
                Tcorr = Ton x exp[((ln(Thps) - ln(Ton) + ln(2))**2)/term2]
            '''

            term1 = (np.log(abs(hps)) - np.log(abs(on)) - np.log(2))**2
            term2 = 4*np.log(2)
            term3 = term1/term2
            pc = np.exp(term3)
            Tcorr = on * pc

            #calculate the error
            if hps < 0.0:
                pf1 = pc * 2.0 * term3 * (-1.0/hps)
            else:
                pf1 = pc * 2.0 * term3 * (1.0/hps)
                                
            pf2 = pc * 2.0 * term3 * (1.0/on)
            Tcorerr = np.sqrt((erron**2)*(pf2**2) + (errhps**2)*(pf1**2))
            
        #if no HPS
        elif((hpn != 0) and (hps== 0)):
            '''
                Tcorr = Ton x pc

                term1 = ((ln(Ton) - ln(Thpn) - ln(2))**2)
                term2 = (4*ln(2))
                term3 = term1/term2
                pc = exp[term3]
            '''

            term1 = (np.log(abs(on)) - np.log(abs(hpn)) - np.log(2))**2
            term2 = 4*np.log(2)
            term3 = term1/term2
            pc = np.exp(term3)
            Tcorr = on * pc

            #calculate the error
            if hpn < 0.0:
                pf1 = pc * 2.0 * term3 * (-1.0/hpn)
            else:
                pf1 = pc * 2.0 * term3 * (1.0/hpn)

            pf2 = pc * 2.0 * term3 * (1.0/on)
            Tcorerr = np.sqrt((erron**2)*(pf2**2) + (errhpn**2)*(pf1**2))


        #if all present
        elif((hpn != 0) and (hps!= 0)):
            '''
                Tcorr = Ton x exp[((ln(Thps) - ln(Thpn))**2)/(16*ln(2))]
            '''
            '''
                Tcorr = Ton x pc

                term1 = ((ln(Thps) - ln(Thpn))**2)
                term3 = term1/term2
                pc = exp[term3]
            '''

            term1 = (np.log(abs(hps)) - np.log(abs(hpn)))**2
            term2 = 16*np.log(2)
            term3 = term1/term2
            pc = np.exp(term3)
            #pcerr=== negligable, why???
            Tcorr = abs(on * pc)

            #calculate the error
            '''if hps < 0.0:
                pf1 = pc * 2.0 * term3 * (1.0/hps)
            else:'''
            pf1 = pc * 2.0 * term3 * (1.0/abs(hps))
                                
            '''if hpn < 0.0:
                pf2 = pc * 2.0 * term3 * (-1.0/hpn)
            else:'''
            pf2 = pc * 2.0 * term3 * (1.0/abs(hpn))


            Tcorerr = np.sqrt((errhpn**2)*(pf2**2) + (errhps**2)*(pf1**2))

        #if no hpn/hps
        if((hpn == 0) and (hps== 0)):
            '''
                Tcorr = 0
            '''

            pc=0
            Tcorr=0
            Tcorerr=0
    else:
        pc=0
        Tcorr=0
        Tcorerr=0
            
    return pc, Tcorr, Tcorerr
```

```python
# Calculate Pointing
pcl, Tcorrl, Tcorerrl = calcPC(max(fitpnl), nlrms, max(fitpsl), slrms, max(fitponl), onlrms)
pcr, Tcorrr, Tcorerrr = calcPC(max(fitpnr), nrrms, max(fitpsr), srrms, max(fitponr), sonrrms)

print 'Amplitude correction factor due to pointing error in  LCP = %.3f ' %(pcl)
print 'Amplitude correction factor due to pointing error in RCP = %.3f ' %(pcr)
print 'Ta_LCP corr = %.3f +- %.3f [K]' %(Tcorrl, Tcorerrl)
print 'Ta_RCP corr = %.3f +- %.3f [K]' %(Tcorrr, Tcorerrr)



```

# 7. Calculate source flux


```python
psslcp = 6.06
errpsslcp = 0.81
pssrcp = 5.71
errpssrcp = 0.73

Slcp = psslcp*Tcorrl
Srcp = pssrcp*Tcorrr

Stot = (Slcp+Srcp)/2

```

```python
print('Flux density = %.3f Jy' %Stot)
```
